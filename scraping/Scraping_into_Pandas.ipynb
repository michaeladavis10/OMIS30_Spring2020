```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)")
```

```python
soup = BeautifulSoup(r.text, "html.parser")
```

```python
main_table = soup.find('table', id = 'main')
```

```python
headers_footers = main_table.find_all('th')
headers = []
for header in headers_footers[0:6]:
    headers.append(header.text)
#     print(header.text)
#     print('\n')
headers
```

```python
rows = main_table.find_all('tr')[1:]
master_list = []
for row in rows:
    country_list = []
    for i in row.find_all('td'):
        print(i.text)
        country_list.append(i.text)
    master_list.append(country_list)
```
