```python
soup = BeautifulSoup(r.text, "html.parser")
```

```python
links = soup.find_all("a")
links
```

```python
divs = soup.find_all('div')
divs
```

```python
tables = soup.find_all('table')
tables
```

```python
main_table = soup.find('table', id = 'main')
main_table
```

```python
headers_footers = main_table.find_all('th')
```

```python
for header in headers_footers[0:6]:
    print(header)
    print('\n')
```

```python
headers = []
for header in headers_footers[0:6]:
    headers.append(header.text)
#     print(header.text)
#     print('\n')
headers
```

```python
rows = main_table.find_all('tr')[1:]
```

```python
for row in rows:
    print(row)
    print('\n')
```

```python
rows[0]
```

```python
for i in rows[0].find_all('td'):
    print(i)
    print('\n')
```

```python
for i in rows[0].find_all('td'):
    print(i.text)
    print('\n')
```

```python
master_list = []
for row in rows:
    country_list = []
    for i in row.find_all('td'):
        print(i.text)
        country_list.append(i.text)
    master_list.append(country_list)
```
